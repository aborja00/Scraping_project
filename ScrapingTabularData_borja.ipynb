```python
import pandas as pd 
```

```python
## Link for OAG website
url= "https://www.atg.wa.gov/data-breach-notifications?page=0"
```

```python
print(url)
```

```python
## Getting all of the pages that have tabular data from the site
base_url= "https://www.atg.wa.gov/data-breach-notifications?page="

urls_fl = []

for url_number in range (0,28):
    urls_fl.append(f"{base_url}{url_number}")
```

```python
## all 28 pages 
urls_fl
```

```python
from random import randrange
import time
```

```python
## total links on the site that will be scraped
total_links= len(urls_fl)
total_links
```

```python
df_all = [] ## where all data will be saved
problem_urls = [] ## if any URLS do not work or refuse to let me scrape data
counter = 0
total_links = len(urls_fl)  # Add this line to define total_links

for url in urls_fl:
    counter += 1
    print(f"Scraping link {counter} of {total_links}")
    try:
        df = pd.read_html(url)
        if df:  # Check if df is not empty
            print(df[0])
            df_all.append(df[0])
        else: ## in case no table is available
            print(f"No tables found at {url}")
            problem_urls.append(url)
    except Exception as e: ## identify any problem areas
        print(f"Oh oh ... problem with {url}: {e}")
        problem_urls.append(url)
    finally:
        snoozer = randrange(5, 12)  # snoozer to make sure the site doesn't block the scraper.  
        print(f"Snoozing for {snoozer} seconds before next scrape")
        time.sleep(snoozer)

print("Done scraping!")
```

```python
#concat all scraped data 
data_breach=pd.concat(df_all, ignore_index=True)
```
